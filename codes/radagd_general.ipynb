### Utils

```python
def enforce_spd(S, eps=1e-19):
    """Project a symmetric matrix to SPD by eigenvalue flooring: numerical stability"""
    S = 0.5 * (S + S.T)
    w, V = np.linalg.eigh(S)                 # symmetric eigendecomp is most stable
    # scale-aware floor: keeps conditioning sane without over-regularizing
    # floor = max(eps, 1e-12 * max(1.0, np.max(np.abs(w))))
    w_clipped = np.clip(w, eps, None)
    return (V * w_clipped) @ V.T
```

## Tasks on manifolds

Target functions

```python
def logdet_ftn_grad(spd_mfd, current_point, target_value, power = 2):
    inside = (np.linalg.slogdet(current_point)[1] - target_value)
    ftn_value = inside ** power
    euclidean_grad = (power * (inside ** (p-1))) * np.linalg.inv(current_point)
    ai_grad = spd_mfd.euclidean_to_riemannian_gradient(current_point, euclidean_grad)
    return ftn_value, ai_grad

def ftn_grad_orcale_logdet(target_value, power):
    def ftn_grad_orcale(mfd, x):
        return logdet_ftn_grad(mfd, x, target_value, power)
    return ftn_grad_orcale
```

```python
def ent_linear_ftn_grad(spd_mfd, current_point, C):
    ftn_value = np.trace(current_point @ C) - np.linalg.slogdet(current_point)[1]
    euclidean_grad = C - np.linalg.inv(current_point)
    ai_grad = spd_mfd.euclidean_to_riemannian_gradient(current_point, euclidean_grad)
    return ftn_value, ai_grad

def ftn_grad_orcale_ent_lin(C):
    def ftn_grad_orcale(mfd, x):
        return ent_linear_ftn_grad(mfd, x, C)
    return ftn_grad_orcale
```

```python
def rayleigh_ftn_grad(mfd, current_point, H):
    hx = np.dot(H, current_point)
    grad = mfd.euclidean_to_riemannian_gradient(current_point, -hx)
    ftn = -0.5 * np.dot(current_point, hx)
    return ftn, grad

def ftn_grad_orcale_rayleigh(H):
    def ftn_grad_orcale(mfd, x):
        return rayleigh_ftn_grad(mfd, x, H)
    return ftn_grad_orcale
```

### RAdaGD class

```python
def radagds(mfd, ftn_grad_orcale, num_iter, step_size=1, L = 1, 
            init_x = None, adaptive = True, A_order = 0, 
            zeta = None, ratio = 1-1e-03, spd = False):
  """
  Conduct GD for f(X) = |\log \det X - target|^p
  """
  # create a list to save the result
  gd_f_list = []

  # initialize
  if init_x is None:
    gd_x_k = mfd.random_point()
  else:
    gd_x_k = init_x

  # save the initial result
  # current_gd_f = -np.linalg.slogdet(gd_x_k)[1]
  current_gd_f, _ = ftn_grad_orcale(mfd, gd_x_k)
  gd_f_list.append(current_gd_f)

  # create a step-size list
  stepsize_list = []

  # For adaptive methods
  list_len = num_iter + 2
  A = [None] * (list_len)
  B = [None] * (list_len)
  tilde_B = [None] * (list_len)
  for i in range(list_len):
    if A_order == 1:
      A[i] = (i+5)/2
      if zeta is None:
        tilde_B[i] = (i+1)/2
        B[i] = ratio * (i+1)/2
      else:
        B[i] = (i+1)/2
        tilde_B[i] = B[i] + zeta
    elif A_order == 0.5:
      a = 2
      A[i] = 2 * np.sqrt(i+a) + 2
      if zeta is None:
        B[i] = ratio * (2 * np.sqrt(i+a-2) - 2 + (5 / (1 + 2 * np.sqrt(i+a-2))))
        tilde_B[i] = 2 * np.sqrt(i+a-2) - 2 + (5 / (1 + 2 * np.sqrt(i+a-2)))
      else:
        B[i] = (2 * np.sqrt(i+a-2) - 2 + (5 / (1 + 2 * np.sqrt(i+a-2))))
        tilde_B[i] = B[i] + zeta
    elif A_order == 0:
      A[i] = 3
      B[i] = 5/4
      tilde_B[i] = 5/4 + 1
    else:
      a = 2
      A[i] = 2 * np.power(i+a, A_order) + 2
      if zeta is None:
        tilde_B[i] = A[i] - 2
        B[i] = ratio * tilde_B[i]
      else:
        B[i] = A[i] - 2
        tilde_B[i] = B[i] + zeta
  s0 = step_size
  s_t1 = s0
  L_e = L
  theta_t = 0
  s_prev = s_t1

  # run the gradient methods
  for curr_iter in range(2, num_iter + 1):
    # calculate gradient
    f_old, gd_grad = ftn_grad_orcale(mfd, gd_x_k)
    # compute adaptive stepsize
    if adaptive:
        t= curr_iter
        s_prev = s_t1
        if t == 0:
          s_t1 = min(min(1/A[0], A[0]/tilde_B[1]) * s_prev, ( A[0]**2 + B[0] * (tilde_B[1]) ) * 1/L_e)
          # s_t1 = min( 1/A[0] * s_prev, A[0] * B[0] / ( A[0]**2 + B[0] * (tilde_B[1]) ) * 1/L_e)
        elif t >= 1:
          t = t+1
          r_s = min((A[t-2]+1)/A[t-1], A[t-1]/tilde_B[t])
          if L_e > 0:
            if s_prev >= (1/L_e):
              r_l = A[t-1] * B[t-1] / ( A[t-1]**2 + B[t-1] * tilde_B[t] )
            else:
              r_l = A[t-1]/(tilde_B[t])
            s_t1 = min(r_s * s_prev, r_l * 1/L_e)
          else:
            s_t1 = r_s * s_prev

        # adaptive GD update
        stepsize_list.append(s_t1)
        gd_x_next = mfd.exp(gd_x_k, -s_t1 * gd_grad)
        # f_new = -np.linalg.slogdet(gd_x_next)[1]
        f_new, gd_grad_next = ftn_grad_orcale(mfd, gd_x_next)
        # compute adaptive stepsize
        gd_grad_next_on_old = mfd.transport(gd_x_next, gd_x_k, gd_grad_next)
        diff = gd_grad_next_on_old - gd_grad
        numerator = mfd.norm(gd_x_k, diff) ** 2
        if numerator == 0:
          L_e = 0
        else:
          denominator = -2*(f_new - f_old + s_t1 * mfd.inner_product(gd_x_k, gd_grad, gd_grad_next_on_old))
          L_e = np.exp(np.log(numerator) - np.log(denominator))
          if L_e < 0 or (math.isnan(L_e)):
            L_e = 0
        gd_x_k = gd_x_next
    else:
      stepsize_list.append(step_size)
      gd_x_k = mfd.exp(gd_x_k, -step_size * gd_grad)
    # save the result
    # numerical stability
    if spd:
      gd_x_k = enforce_spd((gd_x_k))
    current_gd_f, _ = ftn_grad_orcale(mfd, gd_x_k)
    gd_f_list.append(current_gd_f)

  # save the last result
  gd_x_last = gd_x_k

  # return the result
  return gd_f_list, gd_x_last, stepsize_list
```

```python
# Reference: https://github.com/jungbinkim1/RNAG?tab=readme-ov-file

np.random.seed(1234)
d = 1000 #problem dimension
mfd = Sphere(d)
A = np.random.randn(d,d)/np.sqrt(d)
H = (A+A.T)/2
ftn_grad_orcale = ftn_grad_orcale_rayleigh(H)

eigenvalues, eigenvectors = np.linalg.eig(H)
max_eig = np.max(eigenvalues)
min_eig = np.min(eigenvalues)
print('Computing numerical solution...')
x_sol = eigenvectors[:, np.argmax(eigenvalues)]/np.linalg.norm(eigenvectors[:, np.argmax(eigenvalues)])
f_true = -0.5*np.dot(x_sol ,np.dot(H, x_sol))
L = max_eig - min_eig # L-smoothness parameter
init_value = mfd.random_point()
gd_x_last_j = init_value
adap_gd_x_last_j = init_value
```

```python
num_iter = 1000
A_order_experi = 0.1
gd_f_list, gd_x_last, stepsize_gd = radagds(mfd, ftn_grad_orcale, num_iter, step_size=1/L, L=L,
                                            init_x = init_value, adaptive=False)
print("GD Done")
adap_gd_f_list_a01, adap_gd_x_last_a01, stepsize_a01 = radagds(mfd, ftn_grad_orcale, num_iter, 
                                                              step_size=1/L, L=L,
                                                              init_x = init_value, adaptive=True, 
                                                              A_order = A_order_experi,
                                                              zeta = 1)
print(f"AdaGDO({A_order_experi}) Done")
```
